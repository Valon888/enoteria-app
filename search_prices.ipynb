```python
import os

search_terms = ["29.99", "49.99"]
root_dir = os.getcwd()
exclude_dirs = {'.git', 'vendor', 'node_modules', '.idea', '.vscode', 'storage'}
found_matches = []

print(f"Searching for {search_terms} in {root_dir}...")

for root, dirs, files in os.walk(root_dir):
    # Modify dirs in-place to skip excluded directories
    dirs[:] = [d for d in dirs if d not in exclude_dirs]
    
    for file in files:
        if file.endswith(".ipynb"): continue
        if file.endswith(".log"): continue
        if file.endswith(".png"): continue
        if file.endswith(".jpg"): continue
        if file.endswith(".jpeg"): continue
        if file.endswith(".gif"): continue
        if file.endswith(".pdf"): continue
        if file.endswith(".zip"): continue
        
        file_path = os.path.join(root, file)
        try:
            with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
                lines = f.readlines()
                for i, line in enumerate(lines):
                    for term in search_terms:
                        if term in line:
                            # Store relative path for cleaner output
                            rel_path = os.path.relpath(file_path, root_dir)
                            found_matches.append((rel_path, i + 1, term, line.strip()))
        except Exception as e:
            pass

if found_matches:
    print(f"\nFound {len(found_matches)} matches:")
    for match in found_matches:
        print(f"File: {match[0]}, Line: {match[1]}, Term: {match[2]}")
        print(f"  Context: {match[3][:100]}...") # Print first 100 chars of line
else:
    print("\nNo matches found.")
```
