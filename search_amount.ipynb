```python
import os

search_term = "amount"
root_dir = "."

print(f"Searching for '{search_term}' in {os.path.abspath(root_dir)}...")

found_files = []

for dirpath, dirnames, filenames in os.walk(root_dir):
    # Skip .git and other hidden directories
    if '.git' in dirnames:
        dirnames.remove('.git')
    if 'vendor' in dirnames:
        dirnames.remove('vendor')
    if 'node_modules' in dirnames:
        dirnames.remove('node_modules')
        
    for filename in filenames:
        filepath = os.path.join(dirpath, filename)
        try:
            with open(filepath, 'r', encoding='utf-8', errors='ignore') as f:
                lines = f.readlines()
                for i, line in enumerate(lines):
                    if search_term in line:
                        print(f"Found in {filepath}:{i+1}: {line.strip()}")
                        found_files.append(filepath)
                        break # Stop after first match in file to avoid spam
        except Exception as e:
            # print(f"Could not read {filepath}: {e}")
            pass

print(f"\nSearch complete. Found in {len(found_files)} files.")
```
